### Pandas Visualization

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r"C:\Users\user\Documents\Python Tutorials\Ice Cream Ratings.csv")
df = df.set_index('Date')
df
```

```python
print(plt.style.available)
plt.style.use('classic')
```

```python
df.plot()
```

```python
df.plot(kind = 'bar')
```

```python
df.plot(kind = 'line', subplots = True)
```

```python
df.plot(kind = 'bar', subplots = True)
```

```python
df.plot(kind = 'bar', title = 'Ice Cream Ratings', xlabel = 'Daily Ratings', ylabel = 'Scores')
```

```python
df.plot(kind = 'bar', stacked = True)
```

```python
df['Flavor Rating'].plot(kind = 'bar', stacked = True)
```

```python
df.plot.barh(stacked = True)
```

```python
df.plot.scatter(x =  'Texture Rating', y ='Overall Rating', s = 500, c = 'red')
```

```python
df.plot.hist(bins = 10)
```

```python
df.boxplot()
```

```python
df.plot.area(figsize = (10, 5))
```
